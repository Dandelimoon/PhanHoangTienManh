Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó. 


```python
a = float(input("Nhập chiều dài:"))
b = float(input("Nhập chiều rộng:"))
S = a*b
C = (a+b)*2
print("Diện tích:", S)
print("Chu vi:", C)
```

Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó.  

```python
r = float(input())
S = 3.14 * (r**2)
C = 2 * 3.14 * r
print(f'Diện tích là: {S:.2f}')
print(f'Chu vi là: {C:.2f}')
```

 Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a, b, c = map(float, input().split())
p = (a + b + c) / 2
S = (p * (p - a) * (p - b) * (p - c)) ** 1/2
C = a + b + c
if a > 0 and b > 0 and c > 0:
    if (a + b > c) and (a + c > b) and (b + c > a):
        print("a, b, c là 3 cạnh tam giác")
        if a == b == c:
            print("và là tam giác đều")
        elif (a == b and a != c) or (a == c and a != b) or (b == c and b != a):
            print("tam giác cân")
        elif (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
            print("và là tam giác vuông")
        else:
            print("và là tam giác thường")
        print(f"Diện tích tam giác là: {S:.2f}")
        print(f"Chu vi tam giác là: {C:.2f}")
    else:
        print("Khong la ba canh cua mot tam  giac")

else:
    print("Khong la ba canh cua mot tam  giac")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0) 

```python
a, b, c= map(float, input().split())
if a != 0:
    delta= b**2 - 4 * a * c
    x1= (-b + (delta ** 0.5)) / 2 * a
    x2= (-b - (delta ** 0.5)) / 2 * a
    print(f"{x1:.2f}, {x2:.2f}")
else:
    print("Không phải phương trình bậc 2")
```

Bài 8: Giải hệ phương trình tuyến tính:  
ax + by = m   
cx + dy = n 

```python
a, b, m = map(float, input().split())
c, d, n = map(float, input().split())
if a*d - b*c != 0:
    y= (a*n - c*m)/(a*d - b*c)
    x= (m - b*y)/a
    print(f"x={x:.2f}, y={y:.2f}")
else:
    print("hệ phương trình vô nghiệm")
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết 
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao 
nhiêu giây.  

```python
ts = float(input()) 
th = ts / 3600
tm = ts / 60
t1 = int(ts // 3600) 
t2 = int(((ts / 3600) - t1) *60)
t3 = ts - (t1 * 3600) - (t2 * 60)
print("số giờ:", f"{th:.2f}h")
print("số phút:", f"{tm:.2f}m")
print(f"thời gian làm việc là: {t1}h {t2}m {t3}s")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không? 

```python
x0, y0, r = map(float, input().split())
x, y = map(float, input().split())
if r == ((x -x0) ** 2 + (y - y0) ** 2) ** 0.5:
    print("điểm A(x,y) thuộc đường tròn")
else:
    print("điểm A(x,y) không thuộc đường tròn")
```
